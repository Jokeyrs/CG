## Make Rooms

```python
tile_size = 16
room_size = 64

with open('Rooms/room1.txt', 'w') as file:
    for i in range(room_size):
        for j in range(room_size):
            if i == 0 or i == room_size - 1 or j == 0 or j == room_size - 1:
                file.write('#')
            else:
                file.write('.')
            
        file.write('\n')
```
